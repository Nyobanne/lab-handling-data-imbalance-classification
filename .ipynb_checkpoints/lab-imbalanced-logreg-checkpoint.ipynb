# Feature selection with p-values

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
numerical = pd.read_csv('numerical.csv')
numerical.shape
```

```python
targets = pd.read_csv('target.csv')
targets.head()
```

```python
pd.set_option('display.max_columns',None)
all_data = pd.concat((numerical,targets),axis=1)
all_data.head()
```

```python
all_data.shape
```

```python
donations_data = all_data[all_data['TARGET_B']==1]
X = donations_data.drop(columns=['TARGET_B','TARGET_D'])
y = donations_data['TARGET_D']
```

```python
X_added_constant = sm.add_constant(X) #stats model
X_added_constant
model = sm.OLS(y,X_added_constant).fit()
model.summary()
#coeff is every time feature increases 1, target increases 1*coeff
```

```python
# keep removing the features with the highest P-value, build a new model each time
# and watch the R-squared value. Typically it wil go up first (removal of noise),
# and then start to go down again (removal of information). That is when you should stop.


# Note: this is very similar to what RFE does under the hood
```

```python
# Methods for Feature Selection

# Combination of:
# - looking at correlation between features and target: keep ft with high cor
# - looking at low variance columns: eliminate columns with var below treshold
# - Kbest or RFE: look at best/worst features, use suggestions but check for multicolinearity

#     OR
# - PCA: create new features
```

# Imbalance

NOTE: 
* for downsampling/upsampling/SMOTE we need both the features and the target in one dataframe: need to keep tem together.<br/>
* Rebalancing is only the training set, not on the test set. So we need to recombine features and targets for the train set, then treat imbalance.<br/>
* The code here is just to remind you of HOW we do it, but be careful on what you apply it to: only on train.

```python
# Typically we would do this as a first step:
# trainset = pd.concat([X_train, y_train], axis=1)
# but for this demo we will use:
trainset=all_data #imagine that from here in we are working on training portion of the data
```

```python
trainset['TARGET_B'].value_counts() #17 to 1 imbalance
```

```python
data_B = trainset.drop(['TARGET_D'], axis=1).copy()
data_B.head() #targetB = did the donor donate (boolean)
```

```python
from sklearn.utils import resample

category_0 = data_B[data_B['TARGET_B'] == 0]
category_1 = data_B[data_B['TARGET_B'] == 1]
```

## Downsampling (undersampling)

```python
category_0_undersampled = resample(category_0,
                                   replace=False,
                                   n_samples = len(category_1)) #resample remove majority
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
```

```python
data_downsampled['TARGET_B'].value_counts()
```

## Upsampling (oversampling)

```python
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0)) #resample replace minority
```

```python
print(category_0.shape)
print(category_1_oversampled.shape) #shapes are now the sae
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
data_upsampled['TARGET_B'].value_counts()
```

```python
# and then there is SMOTE: using data that doesnt exist
#Synthetic Minority Oversampling TEchnique => figures out which one is minority. but data is synthetic
```

# Decision Trees

## Classification

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer
```

```python
print(cancer["DESCR"])
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)

#cancer = pd.read_csv('cancer.csv')
#features = cancer.drop('target')
#labels = cancer['target']

features = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
labels = pd.DataFrame(cancer['target'], columns = ['labels'])
display(features.head())
display(labels.head(20))
```

```python
labels.labels.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=2)
model.fit(X_train, y_train)
```

```python
print("test data accuracy was ",model.score(X_test,y_test))
# 100% on training data
print("train data accuracy was ",model.score(X_train,y_train))
```

```python
labels.labels.value_counts()
357/(357+212)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print(list(y_test['labels']))
```

```python
# to find the optimum depth, I repeat the process but choose the max_depth of the tree
max_depth = range(1,30)
test = []
train = []

for depth in max_depth:
    model = DecisionTreeClassifier(max_depth=depth, random_state=0)
    model.fit(X_train, y_train)
    test.append(model.score(X_test,y_test))
    train.append(model.score(X_train,y_train))
```

```python
# we quickly see overfitting properties
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot( train, label="training accuracy")
plt.plot( test, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_depth-1")
plt.legend()
```

```python
# we choose 2 as our optimum tree depth
model = DecisionTreeClassifier(max_depth = 4)
model.fit(X_train, y_train)
```

```python
model.feature_importances_
```

```python
# this code helps you plot the feature importance!
import numpy as np
def plot_feature_importances(model, namelist):
    n_features = len(model.feature_importances_)
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), namelist)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")

plot_feature_importances(model, cancer.feature_names)
```
